## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties',[['info','governor']])
```

```python
# further populate tables created from nested element
# json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])

json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

#### Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
print(data['mjtheme_namecode'][1])
```

```python
df=pd.DataFrame([])
doc=pd.DataFrame([])
for i,item in enumerate(list(data['mjtheme_namecode'])):
    for j in range(len(item)):
        doc=pd.DataFrame.from_dict(item)
        df=df.append(doc)
```

```python
df.head()
```

```python
df.shape
```

```python
df.groupby(['name']).size().sort_values(ascending=False).head(11)
```

```python
filt=df['code']=='11'
df[filt].drop_duplicates()
```

```python
filt=df['name']!=''
lookup=df[filt].drop_duplicates().sort_values(by='code')
lookup.reset_index(inplace=True)
lookup.drop(columns='index')
```

```python
df2=pd.merge(df,lookup,how='left', on='code')

df2.drop(columns=['name_x','index'], inplace=True)
df2.rename(columns={'name_y':'name'},inplace=True)
df2.head(10)
```

#### Top 10 major project themes after filling missing values

```python
df2.groupby(['name']).size().sort_values(ascending=False).head(10)
```
